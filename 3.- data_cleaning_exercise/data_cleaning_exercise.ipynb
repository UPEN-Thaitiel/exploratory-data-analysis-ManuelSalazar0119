```python

import pandas as pd
import numpy as np


data = {
    "Name": ["Gaby", "Jaz", np.nan, "Manuel"],
    "Age": [25, np.nan, 30, 35],
    "Score": [85, 90, np.nan, 88]
}

df = pd.DataFrame(data)

print("DataFrame Original:")
print(df.to_markdown(tablefmt="grid", stralign="center"))
print("\n")


age_mean = round(df['Age'].mean(), 2)
df['Age'] = df['Age'].fillna(age_mean)


score_mean = round(df['Score'].mean(), 2)
df['Score'] = df['Score'].fillna(score_mean)

print("DataFrame después de reemplazar nulos:")
print(df.to_markdown(tablefmt="grid", stralign="center"))
print("\n")

df = df.rename(columns={
    "Name": "Nombre",
    "Age": "Edad",
    "Score": "Puntaje"
})

print("DataFrame con columnas renombradas:")
print(df.to_markdown(tablefmt="grid", stralign="center"))
print("\n")

new_data = {
    "Nombre": ["Eva", "Frank"],
    "Edad": [28, 40],
    "Puntaje": [92, 78]
}
df2 = pd.DataFrame(new_data)

df_concatenado = pd.concat([df, df2], ignore_index=True)

print("DataFrame concatenado:")
print(df_concatenado.to_markdown(tablefmt="grid", stralign="center"))
print("\n")


group_data = {
    "Clust": ["A", "B", "A", "B", "C", "C"],
    "Score": [85, 90, 88, 72, 95, 80],
    "Age": [15, 16, 15, 17, 16, 15]
}

df_group = pd.DataFrame(group_data)

grouped = df_group.groupby('Clust').mean()

print("Agrupamiento por categorías 'Clust':")
print(grouped.to_markdown(tablefmt="grid", stralign="center"))
```
